1.explaint the difference between relative,absolute and mixed cell referencing.you have the following datasets.
vehicle      price
car          500000
bycyle       10500
autoricksaw  25000
answer1.Relative reference
        the reference changes when we move the formula
        .Format :B2
        .working:if we write =B2(car price) and drag it down one cell, it automatically changes to B3(bicycle price). it follows the direction of our movement.
        2.Absolute reference:this locks the reference to one specific cell.
          it never changes, no matter where we copy theformula.
        .Format:$B$2
        .Working: The dollar signs acts as locks. if we use =$B$2 and drag it down every single cell will still point exactly to the car price (500000). This is useful for fixed values like a tax rate or discount percentage.
        3.Mixed reference: This locks either the column or the row , but not both.
        .Format:$B2(column locked) or B$2(row locked).
        .Working: *$B2: if we drag this across columns, it stays on column B. if we drag it down , the row number still changes.
